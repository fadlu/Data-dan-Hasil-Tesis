```python
import pandas as pd
import numpy as np
import pubchempy as pc
import time
```

```python
dataset = pd.read_csv('D:/Thesis/Data/SUPERTARGETDRUGBANK_CLEAN_NEWApproved.csv')
dataset
```

```python
list_cid_unik = dataset['CID'].unique()
list_protein_unik = dataset['Gene Symbol'].unique()
```

```python
compound_smiles = []
compound_fingerprint=[]
j=0
for i in list_cid_unik:
    print("senyawa ke", j) 
    print("cid",i)
    t0 = time.time()
    #Smiles
    compound_smiles.append((i,pc.Compound.from_cid(str(i)).canonical_smiles))
    #Fingerprint
#     fp = [i]; fp.extend([j for j in pc.Compound.from_cid(str(i)).cactvs_fingerprint])
#     compound_fingerprint.append(fp)
    t1 = time.time()
    print("waktu proses", t1-t0)
    j=j+1
com_smiles = pd.DataFrame(compound_smiles, columns = ["CID_senyawa", "smiles"])
# com_fingerprint1 = pd.DataFrame(compound_fingerprint, columns=["CID_senyawa"] + list(range(1,882)))
```

```python
com_smiles = pd.DataFrame(compound_smiles, columns = ["CID_senyawa", "smiles"])
# com_fingerprint1.to_csv('com_fingerprint.csv', index=False)
com_smiles.to_csv('D:\Thesis\Data\com_smiles9.csv', index=False)
```

```python
from PyFingerprint.All_Fingerprint import get_fingerprint
list_fp = ['daylight' ,'klekota-roth','circular']
size_fp = [1024, 4860, 1024]
for i in range(len(list_fp)-2):
    print('Processing for FP:', list_fp[i])
    com_fingerprint_fp = pd.DataFrame(np.zeros((len(com_smiles), size_fp[i])))
    j=0
    tracknumber = 0
    for k in com_smiles['smiles']:
        print("CID senyawa", com_smiles.iloc[j,0])
        print("jumlah senyawa", tracknumber)
        fps = get_fingerprint(k, fp_type=list_fp[i])
        com_fingerprint_fp.loc[j,fps]=1
        j=j+1
        tracknumber=tracknumber+1
#     com_fingerprint_fp = com_fingerprint_fp.astype('int32')
    com_fingerprint_fp.to_csv('D:\Thesis\Data\com_fp_'+list_fp[i] +'NewFiltered.csv', index=False)
#Fingerprint
```
